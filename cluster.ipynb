导入包

```python
import numpy as np
import pandas as pd
from sklearn import metrics
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

环形数据1生成

```python
from sklearn import datasets
n_samples=300
X1,y1=datasets.make_circles(n_samples=n_samples,factor=.5,noise=.05)
```

片形数据2生成

```python
X2,y2=datasets.make_blobs()
```

S形布局数据3生成

```python
X3,y3=datasets.make_s_curve()
```

Kmeans模型建立

```python
# from sklearn.cluster import KMeans
kmeans1=KMeans(n_clusters=2,random_state=0).fit(X1)
klabels1=kmeans1.labels_
kmeans2=KMeans(n_clusters=2,random_state=0).fit(X2)
klabels2=kmeans2.labels_
kmeans3=KMeans(n_clusters=2,random_state=0).fit(X3)
klabels3=kmeans3.labels_
```

DBSCAN模型的建立

```python
# from sklearn.cluster import DBSCAN
Dbscan1=DBSCAN(eps=0.2,min_samples=2,metric='euclidean').fit(X1)
DBlabels1=Dbscan1.labels_
Dbscan2=DBSCAN(eps=0.2,min_samples=2,metric='euclidean').fit(X2)
DBlabels2=Dbscan2.labels_
Dbscan3=DBSCAN(eps=0.2,min_samples=2,metric='euclidean').fit(X3)
DBlabels3=Dbscan3.labels_
```

调整兰德系数

```python
print("1为环形数据，2为片状数据，3为S形数据")
print("kmeans1",metrics.adjusted_rand_score(y1,klabels1))
print("kmeans2",metrics.adjusted_rand_score(y2,klabels2))
print("kmeans3",metrics.adjusted_rand_score(y3,klabels3))
print("Dbscan1",metrics.adjusted_rand_score(y1,DBlabels1))
print("Dbscan2",metrics.adjusted_rand_score(y2,DBlabels2))
print("Dbscan3",metrics.adjusted_rand_score(y3,DBlabels3))
```

同质性

```python
print("1为环形数据，2为片状数据，3为S形数据")
print("kmeans1",metrics.homogeneity_score(y1,klabels1))
print("kmeans2",metrics.homogeneity_score(y2,klabels2))
print("kmeans3",metrics.homogeneity_score(y3,klabels3))
print("Dbscan1",metrics.homogeneity_score(y1,DBlabels1))
print("Dbscan2",metrics.homogeneity_score(y2,DBlabels2))
print("Dbscan3",metrics.homogeneity_score(y3,DBlabels3))
```

完整性

```python
print("1为环形数据，2为片状数据，3为S形数据")
print("kmeans1",metrics.completeness_score(y1,klabels1))
print("kmeans2",metrics.completeness_score(y2,klabels2))
print("kmeans3",metrics.completeness_score(y3,klabels3))
print("Dbscan1",metrics.completeness_score(y1,DBlabels1))
print("Dbscan2",metrics.completeness_score(y2,DBlabels2))
print("Dbscan3",metrics.completeness_score(y3,DBlabels3))
```
